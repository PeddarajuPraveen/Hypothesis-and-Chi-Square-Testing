# Hypotheses Testing

## Step 1: Define Hypotheses
- **Null Hypothesis (H0):**  
  The true mean weekly cost = theoretical mean  

- **Alternative Hypothesis (H1):**  
  The true mean weekly cost > theoretical mean

This is a **right-tailed Z-test**.

## Step 2: Theoretical Mean

```python
# Theoretical mean cost
X_mean = 600
mu_W = 1000 + 5 * X_mean
mu_W
```

## Step 3: Population Standard Deviation & Standard Error

```python
import math

# Given
sigma_X = 25      # std of X
n = 25            # sample size

# Std of W
sigma_W = 5 * sigma_X
# Standard Error
SE = sigma_W / math.sqrt(n)

sigma_W, SE
```

## Step 4: Test Statistic

```python
# Sample mean
x_bar = 3050

# Test Statistic
Z = (x_bar - mu_W) / SE
Z
```

## Step 5: Critical Value
For a **one-tailed Z-test** with Î± = 0.05:  
$$
Z_{critical} = 1.645
$$

```python
from scipy.stats import norm

alpha = 0.05
Z_critical = norm.ppf(1 - alpha)  # 95th percentile
Z_critical
```

## Step 6: Decision

```python
if Z > Z_critical:
    decision = "Reject Null Hypothesis"
else:
    decision = "Fail to Reject Null Hypothesis"

decision
```

## Step 7: Conclusion
- Test Statistic (Z): calculated value  
- Critical Value (Z): 1.645  
- Decision: Fail to reject H0  

**Conclusion:**  
There is no evidence to support the restaurant owners' claim that weekly operating costs are higher.  
In fact, the observed mean cost (3050) is much lower than the predicted mean (4000).
